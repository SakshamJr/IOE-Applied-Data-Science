```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(100,2,n_informative=2)
```

```python
# Model Initialization
# Model Fit
# Model Predict
# Metrics
```

### Model Initialization

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
y_pred = model.predict(X)
```

```python
from sklearn.metrics import (
    mean_absolute_percentage_error,
    mean_squared_error,
    root_mean_squared_error,
    r2_score,
)

print(f"MAPE: {mean_absolute_percentage_error(y,y_pred)}")
print(f"MSE: {mean_squared_error(y,y_pred)}")
print(f"RMSE: {root_mean_squared_error(y,y_pred)}")
print(f"R2: {r2_score(y,y_pred)}")
```

#### Lasso

```python
from sklearn.linear_model import Lasso
model = Lasso()
model.fit(X,y)
y_pred = model.predict(X)

print(f"MAPE: {mean_absolute_percentage_error(y,y_pred)}")
print(f"MSE: {mean_squared_error(y,y_pred)}")
print(f"RMSE: {root_mean_squared_error(y,y_pred)}")
print(f"R2: {r2_score(y,y_pred)}")
```

```python
from sklearn.datasets import make_classification

X, y = make_classification()
X.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,y)
y_pred = model.predict(X)
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```

Accuracy - The most simple, how much % are accurate =  TN+TP/Total
Precision - Out of total predicted true, how much are actually true
Recall - Out of Actual true, how much were predicted true
F1 Score - Trade-off Between Precision and Recall

Example case for Cancer Diagnosis:

High precision - positive bhaneko ma positive nai bhayo, no unnecessary torture to negative patients

High recall - actual cancer lageko manxe lai accurately positive bhanne bhayo

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y, y_pred))
```
